```python
import pandas as pd
import numpy as np
Toronto=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0] #downloading of the data
```

```python
df_toronto=pd.DataFrame(data=Toronto)
df_toronto.rename(columns={'Postal Code':'PostalCode'}, inplace=True) #column name as requested in the assignment
```

```python
df_toronto.set_index('PostalCode', inplace=True)
df=df_toronto[(df_toronto.Borough != 'Not assigned')] #removing rows that were 'not assigned', moving the result to df
df.reset_index()
#result includes neighborhoods on one row if they share PostalCode (no need to change that)
```

```python
df[df['Neighborhood'].str.match('Not')] #checking that no "Not assigned" neighborhoods are left (none are)
```

```python
df.sort_values(by=['PostalCode'], inplace=True) #PostalCodes are unique
```
